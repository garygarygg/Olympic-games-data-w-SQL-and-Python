### Correlation coefficient & p-value

#### Replace inf, -inf values as NaN & select non null rows

```python
dropinfsnans = dropped_duplicate[dropped_duplicate.replace([np.inf, -np.inf], np.nan).notnull().all(axis=1)]
```

#### Look thru the data, grab the numeric features & add the correlations to the matrix

```python
dropinfsnans.corr()
```

#### Correlation between the pair: weight & age

```python
dropinfsnans.Weight.corr(dropinfsnans.Age)
```

#### Next is the correlation, and the p-value for the correlation

```python
from scipy import stats

corr = stats.pearsonr(dropinfsnans.Weight, dropinfsnans.Age)
corr
```

#### Round up the values

```python
r, p = stats.pearsonr(dropinfsnans.Weight, dropinfsnans.Age)
print  (round(r,4))
print  (round(p,4)) #  denotes a VERY small number, basically 0 plus further 176 decimals
```

#### Create a dataframe for the above output

```python
corr_df = pd.DataFrame(columns=['Weight correlation', 'P-Value'])

for col in dropinfsnans:
    if pd.api.types.is_numeric_dtype(dropinfsnans[col]) and col != 'ID':
        r, p = stats.pearsonr(dropinfsnans.Weight, dropinfsnans[col])
        corr_df.loc[col] = [round(r,3), round(p,3)]
    
corr_df
```
